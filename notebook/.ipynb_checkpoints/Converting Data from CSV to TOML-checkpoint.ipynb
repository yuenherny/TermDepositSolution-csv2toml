```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_path = "../dataset/solution.csv"
df = pd.read_csv(dataset_path)
df
```

```python
df["actual_value"] = df["actual_value"].astype("category")
df["actual_value_num"] = df["actual_value"].cat.codes
df
```

```python
unique_value_count = df["actual_value"].value_counts()
unique_value_count
```

```python
unique_value_count.index[0]
```

```python
plt.figure(figsize=(15,10))
plt.pie(unique_value_count[:], labels=unique_value_count.index[:], autopct="%0.01f%%", textprops={'fontsize':20})
plt.show()
```

```python
sol_key_dict = pd.DataFrame(df.iloc[:, 0]).T.to_dict(orient='records')
print(sol_key_dict)
```

```python
sol_val_dict = pd.DataFrame(df.iloc[:, -1]).T.to_dict(orient='records')
print(sol_val_dict)
```

```python
import toml
```

```python
solkey_dict_path = "../toml/solkey.toml"
solval_dict_path = "../toml/solval.toml"

with open(solkey_dict_path, 'w') as f:
    toml.dump(sol_key_dict, f)

with open(solval_dict_path, 'w') as f:
    toml.dump(sol_val_dict, f)
```
